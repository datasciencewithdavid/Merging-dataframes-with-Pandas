## Merging the datasets

```python
#merging 'train' and 'severity_type'
join1 = pd.merge(train,severity_type)

#viewing the merge dataframes
join1
```

```python
#getting the number of dimensions
print("train dimensions:{}".format(train.shape))
print("severity_type dimensions:{}".format(severity_type.shape))
print("join1 dimensions:{}".format(join1.shape))
```

```python
#checking for missing values
join1.isna().sum()
```

```python
#Checking the number of `id` entries from `sesverity_type` in 'id' of `train`
severity_type['id'].isin(train['id']).value_counts()
```

The dataset `train` and `severity_type` has **7381** rows in common

The above uses the default `inner` join in merging the cells, we proceed to use `left` join below:

```python
#using the `left` join to merge `severity_type` and 'train'
train_1 = train.merge(severity_type, how = 'left', left_on='id', right_on='id')
train_1
```

```python
#comparing the too joins using a boolean condition
train_1 == join1
```

```python
#verifying if there exists a 'False' hidden in the join
'False' in (train_1 == join1)
```

The above boolean result reveals to us that as far as this dataset is concerned, the two types join used is appropriate

```python
#merging 'join1' and 'resource_type'
join2 = pd.merge(join1, resource_type)

#viewing the merged dataframes
join2
```

```python
#getting the number of dimensions
print("join1 dimensions:{}".format(join1.shape))
print("resource_type dimensions:{}".format(resource_type.shape))
print("join2 dimensions:{}".format(join2.shape))
```

```python
#checking for missing values
join2.isna().sum()
```

## Dropping (Removing) Duplicate `id`s in the `join4` Dataset

```python
#dropping duplicate records
join4.drop_duplicates(subset= 'id', keep= 'first', inplace = True)
```

```python
#verifying the change
join4
```

Observe that the index of our dataset after dropping (removing) the duplicate ids is unordered, we'll therefore reset the index so as to have an ordered indices and we also need to sort our `id` column in ascending order too

```python
#sorting the dataset
join4.sort_values(by = ['id'], inplace = True)
```

```python
#verifying the change
join4
```

```python
#resetting index
join4.reset_index(drop = True, inplace = False)
```

Our dataset now has a well ordered index, and the  `id` is as well ordered numerically
